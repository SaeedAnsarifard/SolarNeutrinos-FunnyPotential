```python
import numpy as np
import matplotlib.pyplot as plt
import SurvivalProbability as sp

plt.rcParams['text.usetex']=True
plt.style.use('classic')

plt.rcParams['font.size'] = 18

enu = []
survival_probability = []
gs = [0,10,100,1e5]
for i in (gs):
    enus,survival_probabilitys = sp.main(1.,i,33.)
    enu.append(enus)
    survival_probability.append(survival_probabilitys)
```

```python
fig  = plt.figure(figsize=(10,5))
plt.ylim(0.18,0.62)
plt.xlim(0.1,20)
plt.xscale('log')
for i in range(len(gs)):
    plt.plot(enu[i],survival_probability[i],label=str(gs[i]))
    
plt.legend()
```
