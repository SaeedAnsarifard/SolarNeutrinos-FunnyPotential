```python
import numpy as np
import matplotlib.pyplot as plt
import SurvivalProbability as sp

plt.rcParams['text.usetex']=True
plt.style.use('classic')

plt.rcParams['font.size'] = 18



gs  = [0,1e2,1e3,1e5]
ms  = [1.,10]

enu                  = [[],[]]
survival_probability = [[],[]]
for j,m in enumerate (ms):
    for g in (gs):
        enus,survival_probabilitys = sp.main(m,g,33.)
        enu[j].append(enus)
        survival_probability[j].append(survival_probabilitys)
```

```python
fig  = plt.figure(figsize=(15,10))
fig.subplots_adjust(hspace=0.06)
fig.subplots_adjust(wspace=0.08)

labels = [r'$g_s^f g_p^\nu = 0$',r'$g_s^f g_p^\nu = 10^{-29}$',r'$g_s^f g_p^\nu = 10^{-27}$',r'$g_s^f g_p^\nu = 10^{-25}$']
colors = ['k','g','b','r']

ax0 = fig.add_subplot(211)
ax0.set_ylim(0.18,0.62)
ax0.set_xlim(0.1,20)
ax0.set_xscale('log')
ax0.xaxis.set_major_formatter(plt.NullFormatter())
ax0.tick_params(axis='both', which='minor', length=4)
ax0.tick_params(axis='both', which='major', length=5)
ax0.set_ylabel(r'$\bar{P}_{ee}^{B8} $',fontsize=22)

for i in range(len(gs)):
    ax0.plot(enu[0][i],survival_probability[0][i],colors[i],lw=2,label=labels[i])
    
ax0.set_title(r'$m_\phi = R_\odot^{-1} $',pad=-20)   




ax1 = fig.add_subplot(212)
ax1.set_ylim(0.18,0.62)
ax1.set_xlim(0.1,20)
ax1.set_xscale('log')
#ax1.xaxis.set_major_formatter(plt.NullFormatter())
ax1.tick_params(axis='both', which='minor', length=4)
ax1.tick_params(axis='both', which='major', length=5)
ax1.set_ylabel(r'$ \bar{P}_{ee}^{B8} $',fontsize=22)
ax1.set_xlabel(r'$ E_\nu \ [\rm MeV] $',fontsize=22)

for i in range(len(gs)):
    ax1.plot(enu[1][i],survival_probability[1][i],colors[i],lw=2,label=labels[i])
plt.legend(loc='lower left')
ax1.set_title(r'$m_\phi = 10\times R_\odot^{-1} $',pad=-20)


plt.savefig('P_ee.pdf',bbox_inches='tight')
```
